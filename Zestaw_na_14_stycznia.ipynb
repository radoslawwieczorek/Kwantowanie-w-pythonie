### Przykłady:
matplotlib nie rysuje bezpośrednio wykresów funkcji - rysuje wykresy podanych punktów, np.:

```python
# przykład 1: wykres punktów
import numpy as np
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,0,5,2])

plt.show()
```

Ale oczywiście wykres funkcji można zbudować łącząc punkty wykresu:

```python
# przykład: wykres sinusa
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1) # tworzy tablicę liczb od 0 do 10 ze skokiem 0.1, czyli [0,0.1,0.2,...,10]
y = np.sin(x) # liczy wartości funkcji sin na **wszystkich** elementach tablicy x
              # dostajemy zatem tablicę [sin(0),sin(0.1),sin(0.2),...,sin(10)]

plt.plot(x, y)
plt.show()
```

## Rysowanie liczb zespolonych w `matplotlib` na płaszczyźnie Gaussa
Nie da się bezpośrednio rysować liczb zespolonych funkcją biblioteki `matplotlib` - trzeba wcześniej "wydobyć" z niej część rzeczywistą i urojoną (za pomocą `z.real` i `z.imag`).

Do rysowania punktóm można użyć funkcji `plt.scatter(x,y)`.

*Wskazówki:*
> funkcje `plt.xlim([x0,x1])`,  `plt.ylim([y0,y1])` wyznaczają zakresy osi $Ox$ i $Oy$  na wykresie.
> funkcja `plt.gca()` zwraca aktualnie używane "osie" (*Axes*). Komendą `ax = plt.gca()` zapiszemy "osie" do zmiennej `ax`.
>
> Możemy teraz używając różnych metod dla obiektu `ax` (klasy *Axes*) zmieniać wygląd wykresu:
>
> Np. komenda `ax.set_aspect(1)` zmieni proporcie (*aspect ratio*) jednostek na osiach na równomierne (równoważnie można napisać `ax.set_aspect('equal')`)
### Przykład

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter([1,2],[3,0]) # rysuje dwa punkty (o współrzędnych $(1,3)$ i $(2,0)$)
plt.show() # wyświetla wykres

plt.scatter([1,2],[3,0]) # rysuje dwa punkty (o współrzędnych $(1,3)$ i $(2,0)$)

ax = plt.gca()           # pobiera current Axes (aktualne osie)
ax.set_aspect(1)         # ustawia proporcje na równomierne
ax.grid()                # rysuje siatkę na wykresie

plt.show()
```

### Wektory w Matplotlib
W matplotlib nie ma funkcji rysującej wektor w postaci strzałki. Można narysować odcinek łączący dwa punkty (komendą `plt.plot()`), ale nie będzie zaznaczony zwrot.

Można użyć funkcji `plt.quiver()` (zasadniczo przesnaczonej do rysowania pól wektorowych) wymuszając na niej zachowanie, którego oczekujemy.

>funkcja `plt.quiver(x0, y0, vx, vy, angles='xy', scale_units='xy', scale=1)` narysuje wektor o początku w $(x0,y0)$ i wartości  $(vx,vy)$ - to znaczy, o końcu w punkcie $(x0+vx, y0+vy)$.

### Przykład.


```python
import matplotlib.pyplot as plt

plt.quiver(1, -0.5, 3, 1, color = "black", angles='xy', scale_units='xy', scale=1)
# musimy "ręcznie" zmienić zakres osi, żeby zobaczyć cały wektor:
plt.xlim([0, 5])
plt.ylim([-1, 2])

plt.show()
```
